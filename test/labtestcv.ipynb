```python
#Pillow - Библиотека обработки изображений Python предоставляет несколько методов для извлечения информации, относящейся к
#изображению. Метод histogram () предоставляет информацию о количестве разных цветов / полос.
#Если коротко, то с помощью подключения Pillow мы можем считывать RGB

#Метод histogram () возвращает список количества пикселей для каждой полосы, присутствующей в изображении. 
#В списке будут объединены все счетчики для каждой полосы.

#Если изображение находится в режиме «RGB», то для каждой полосы / цвета будет возвращен список счетчиков пикселей, всего 768.

#Другими словами, для изображения RGB метод histogram () дает информацию о том, 
#сколько видов красных, зеленых и синих пикселей присутствует в изображении 
#для каждых 256 типов красного, 256 типов зеленого и 256 типов синего.

#Построить гистограмму по каждому из каналов (RGB)

from PIL import Image
import matplotlib.pyplot as plt

def getRed(redVal):
    return '#%02x%02x%02x' % (redVal, 0, 0)

def getGreen(greenVal):
    return '#%02x%02x%02x' % (0, greenVal, 0)

def getBlue(blueVal):
    return '#%02x%02x%02x' % (0, 0, blueVal)
```

```python
# создать изображение с определенным значением RGB

from google.colab import files
from io import BytesIO
from skimage import io

uploaded = files.upload()
image = Image.open(BytesIO(uploaded['image.png']))

# Изменение цвета двух пикселей
#image.putpixel((0,1), (1,1,5))
#image.putpixel((0,2), (2,1,5))

# Показать изображение
plt.imshow(image)
```

```python
# Получите цветную гистограмму изображения
histogram = image.histogram()

# Возьмём только красные пиксели
l1 = histogram[0:256]
# Возьмём только синие пиксель
l2 = histogram[256:512]
# Возьмём только зеленый пиксель
l3 = histogram[512:768]

plt.figure(0)

# R гистограмма
for i in range(0, 256):
    plt.bar(i, l1[i], color = getRed(i), edgecolor=getRed(i), alpha=0.3)

# G гистограмма
plt.figure(1)
for i in range(0, 256):
    plt.bar(i, l2[i], color = getGreen(i), edgecolor=getGreen(i),alpha=0.3)

# B гистограмма
plt.figure(2)
for i in range(0, 256):
    plt.bar(i, l3[i], color = getBlue(i), edgecolor=getBlue(i),alpha=0.3)

plt.show()
```
