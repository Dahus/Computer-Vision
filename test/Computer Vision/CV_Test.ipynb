```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
image = cv.imread("tesla.jpg")
```

```python
plt.hist(image.ravel(), 256, [0, 256])
plt.show()
```

```python
colors = ('blue', 'green', 'red')
for i, color in enumerate(colors):
  hist = cv.calcHist([image], [i], None, [256], [0, 256])
  plt.plot(hist, color=color)
  plt.xlim([0, 256])
plt.show()
```

```python
import PIL
from PIL import Image
```

```python
image = cv.imread("tesla.jpg")
```

```python
width, height = image.size
```

```python
colors = image.getcolors(width*height)
```

```python
def hexencode(rgb):
  r = rgb[0]
  g = rgb[1]
  b = rgb[2]
  return '#%02x%02x%02x' % (r,g,b)
```

```python
for index, color in enumerate(colors):
  plt.bar(index, color[0], color=hexencode(color[1]), edgecolor=hexencode(color[1]))
plt.show()
```

```python
size = 256
```

```python
r = np.zeros(size, dtype=int)
g = np.zeros(size, dtype=int)
b = np.zeros(size, dtype=int)
```

```python
for i in image:
  for j in i:
    r[j[2]] += 1
    g[j[1]] += 1
    b[j[0]] += 1
```

```python
plt.plot(r, color="red")
plt.plot(g, color="green")
plt.plot(b, color="blue")
plt.show()
```
